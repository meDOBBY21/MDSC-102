```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
```

```python
col_names=['class',
           'Alcohol',
           'Malic acid',
           'Ash',
           'Alcalinity of ash',
           'Magnesium',
           'Total phenols',
           'Flavanoids',
           'Nonflavanoid phenols',
           'Proanthocyanins',
           'Color intensity',
           'Hue',
           'OD280/OD315 of diluted wines',
           'Proline']
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/wine/wine.data',names=col_names)
df
```

```python
sns.countplot(x='class',data=df)
```

Using Count plot we can see that class 2 has more data points

```python
sns.relplot(x='Alcohol',y='Magnesium',data=df,size='class',hue='class')
```

```python
# @title Scatter plot with respective 'Alcohol'
fig, axes = plt.subplots(4,3,figsize=(15,15))
i,j=0,0
for col in df.columns:
    if col!='class' and col!='Alcohol':
        sns.scatterplot(x='Alcohol',y=col,data=df,hue='class',ax=axes[i,j],palette='deep')
        j+=1
        if j>=3:
            i+=1
            j=0
```
