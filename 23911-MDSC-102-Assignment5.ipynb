#Histogram

---



```python
def histplot(df,col):
    class1=df[df['class']==1]
    class2=df[df['class']==2]
    class3=df[df['class']==3]

    fig,axes = plt.subplots(1,3,figsize=(15,5))
    fig.suptitle(col)
    sns.histplot(data=class1,x=col,hue='class', kde=True,ax=axes[0],palette='dark')
    sns.histplot(data=class2,x=col,hue='class', kde=True,ax=axes[1],palette='dark')
    sns.histplot(data=class3,x=col,hue='class', kde=True,ax=axes[2],palette='dark')

    plt.show()
```

```python
for col in df.columns:
    if col!='class' and col!='Alcohol':
        histplot(df,col)
```
