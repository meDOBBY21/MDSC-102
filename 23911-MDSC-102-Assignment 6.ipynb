```python
def skewPlot(df,col):
    fig,axes=plt.subplots(1,3,figsize=(13,2))

    sns.boxplot(x=col,data=df,ax=axes[0],color='r')
    sns.histplot(x=col,data=df,kde=True,ax=axes[1],color='g',label='Skew:'+str(round(df[col].skew(),3)))
    res=stats.probplot(df[col], dist=stats.loggamma, sparams=(2.5,), plot=axes[2])


    axes[1].legend()
    plt.show()
```

```python
for col in df:
    if col!='class':
        skewPlot(df,col)
```

#Tranforming

---



```python
def transform(df):
    transformed={}
    for col in df:
        if col!='class':
            y, lam = stats.boxcox(df[col])
            transformed[col]=(df[col]**lam-1)/lam

    tdf=pd.DataFrame(transformed)
    return tdf
```

```python
tdf=transform(df)
```

```python
for col in tdf:
    if col!='class':
        skewPlot(tdf,col)
```
