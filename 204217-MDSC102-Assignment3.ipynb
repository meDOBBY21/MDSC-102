#**Plotting the Special Distribution Functions**

Import statements

```python
import numpy as np
from math import comb,factorial,sqrt,pi
import matplotlib.pyplot as plt
```

#Binomial Distribution


$$B(n,p) = \sum_{j=0}^{x} \binom{n}{j} p^j (1 - p)^{n - j}$$

Binomial Distribution function

```python
def B(n,p):
    X,Y=[],[]
    for x in range(n+1):
        X.append(x)
        y=comb(n,x)*(p**x)*((1-p)**(n-x))
        Y.append(y)
    return X,Y
```

Plot

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(10,10))

X,Y=B(20,0.5)
axes[0,0].set_xlabel('X Values')
axes[0,0].set_ylabel('Function Values')
axes[0,0].set_title('BINOMIAL DISTRIBUTION(20,0.5)')
axes[0,0].bar(X,Y)


X,Y=B(20,.1)
axes[0,1].set_xlabel('X Values')
axes[0,1].set_ylabel('Function Values')
axes[0,1].set_title('BINOMIAL DISTRIBUTION(20,0.1)')
axes[0,1].bar(X,Y)


X,Y=B(30,.05)
axes[1,0].set_xlabel('X Values')
axes[1,0].set_ylabel('Function Values')
axes[1,0].set_title('BINOMIAL DISTRIBUTION(30,0.05)')
axes[1,0].bar(X,Y)


X,Y=B(25,.25)
axes[1,1].set_xlabel('X Values')
axes[1,1].set_ylabel('Function Values')
axes[1,1].set_title('BINOMIAL DISTRIBUTION(25,0.25)')
axes[1,1].bar(X,Y)

plt.show()
```

#Poisson Distribution

$$Pois(λ)=e^{-λ}\frac{λ^x }{x!}$$



Poisson Function

```python
def pois(lam):
    X,Y=[],[]
    for x in range(20):
        X.append(x)
        y=np.exp(-1*lam)*(lam**x)/factorial(x)
        Y.append(y)
    return X,Y
```

Plot

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(10,10))

X,Y=pois(10)
axes[0,0].set_xlabel('X Values')
axes[0,0].set_ylabel('Function Values')
axes[0,0].set_title('POISSION DISTRIBUTION(10)')
axes[0,0].bar(X,Y)

X,Y=pois(5)
axes[0,1].set_xlabel('X Values')
axes[0,1].set_ylabel('Function Values')
axes[0,1].set_title('POISSION DISTRIBUTION(5)')
axes[0,1].bar(X,Y)

X,Y=pois(20)
axes[1,0].set_xlabel('X Values')
axes[1,0].set_ylabel('Function Values')
axes[1,0].set_title('POISSION DISTRIBUTION(20)')
axes[1,0].bar(X,Y)

X,Y=pois(15)
axes[1,1].set_xlabel('X Values')
axes[1,1].set_ylabel('Function Values')
axes[1,1].set_title('POISSION DISTRIBUTION(15)')
axes[1,1].bar(X,Y)

plt.show()
```

#Exponential Distribution

$$Exp(λ)=λe^{-λx}$$

Exponential Function

```python
def Exp(lam):
    X=np.linspace(0,10,50)
    Y=[]
    for x in X:
        y=lam*np.exp(-1*lam*x)
        Y.append(y)
    return X,Y
```

Plot

```python
plt.xlabel('X values')
plt.ylabel('Function Values')
plt.title('EXPONENTIAL DISTRIBUTION')

X,Y=Exp(1)
plt.plot(X,Y,label='Exp(1)')

X,Y=Exp(2)
plt.plot(X,Y,label='Exp(2)')

X,Y=Exp(.5)
plt.plot(X,Y,label='Exp(0.5)')

X,Y=Exp(.25)
plt.plot(X,Y,label='Exp(0.25)')

plt.legend()
plt.grid()
plt.show()
```

#Normal Distribution

$$N(𝞵,σ^2)=\frac{1}{\sqrt{2\pi}σ}e^{-(x-𝞵)^2/2σ^2}$$

Normal Distribution

```python
n=lambda mean,var,x:(1/np.sqrt(2*pi*var))*np.exp((-1/2)*(((x-mean)**2)/var))
def N(mean,var):
    X=np.linspace(-20,20,100)
    Y=[]
    for x in X:
        y=n(mean,var,x)
        Y.append(y)
    return X,Y
```

Plot

```python
plt.xlabel('X Values')
plt.ylabel('Function Values')
plt.title('NORMAL DISTRIBUTION')

X,Y=N(0,20)
plt.plot(X,Y,label='N(0,20)')
plt.stem(0,n(0,20,0))

X,Y=N(1,10)
plt.plot(X,Y,label='N(1,10)',color='orange')
plt.stem(1,n(1,10,1),linefmt='orange')

X,Y=N(2,30)
plt.plot(X,Y,label='N(2,30)',color='g')
plt.stem(2,n(2,30,2),linefmt='g')

X,Y=N(0,50)
plt.plot(X,Y,label='N(0,50)',color='r')
plt.stem(0,n(0,50,0),linefmt='r')

plt.grid()
plt.legend()

plt.show()
```

#Gamma Distribution

$$F(α,λ)=\frac{λ^α}{Γ(α)}x^{α-1}e^{-λx}$$

$$Γ(α)=(α-1)!$$

```python
def gammaPDF(alpha,lam):
    X=np.linspace(0,20,200)
    Y=[]
    for x in X:
        y=(lam**alpha*x**(alpha-1)*np.exp(-1*lam*x))/factorial(alpha-1)
        Y.append(y)
    return X,Y
```

```python
plt.xlabel('X Values')
plt.ylabel('Y values')
plt.title('GAMMA DISTRIBUTION')



X,Y=gammaPDF(2,2)
plt.plot(X,Y,label='Gamma(2,2)')

X,Y=gammaPDF(3,2,)
plt.plot(X,Y,label='Gamma(3,2)')

X,Y=gammaPDF(5,1)
plt.plot(X,Y,label='Gamma(5,1)')

X,Y=gammaPDF(9,.5)
plt.plot(X,Y,label='Gamma(9,0.5)')

plt.grid()
plt.legend()
plt.show()
```
